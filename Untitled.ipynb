```python

def make_simple_graphs(name,args):
    conn=[]
    qneg=[]
    qpos=[]
    id=name
    #print(args)
    i=0
    for x in args:
        
        if int(x)<0:
        	#print(x)
        	qneg.append([x,id[i]])
        elif int(x)>=0:
        	#print(x)
        	qpos.append([x,id[i]])
        i+=1

    j=0
    k=0
    # print("qn:",qneg)
    # print("qp:",qpos)
    special=0
    
    while k<len(qneg):
        #pdb.set_trace()
        if abs(qneg[k][0])<qpos[j][0]:
            qpos[j][0]+=qneg[k][0]
            conn.append([qneg[k][1],qpos[j][1],-qneg[k][0]])
            qneg[k][0]=0
            
        elif abs(qneg[k][0])> qpos[j][0]:
            qneg[k][0]+=qpos[j][0]
            conn.append([qneg[k][1],qpos[j][1],qpos[j][0]])
            qpos[j][0]=0
        else:
            conn.append([qneg[k][1],qpos[j][1],qpos[j][0]])
            qpos[j][0]=0
            qneg[k][0]=0
            special=1
            j+=1
            k+=1
        if special==0:
            if qpos[j][0]==0:
                j+=1
                
            elif qneg[k][0]==0:
                k+=1

    #print(conn)
    return [tuple(x) for x in conn]

#make_simple_graphs(['a','b','c','d','e','f'],-40,52,-117,-95,81,119)    
```
